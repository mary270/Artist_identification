```
#Importing libraries
from skimage.io import imread, ImageCollection
import matplotlib.pyplot as plt
import numpy as np
from skimage.color import rgb2gray
```

```

# loading dataset of 20 images
ic = ImageCollection('/content/drive/My Drive/Datasets/resized/resized/*.jpg')
```

```
fig, axes = plt.subplots(100, 5, figsize=(25, 25))
ax = axes.ravel()

for a in ax:
  a.axis('off')
  
grayscale_imags = []
for i,img in enumerate(ic):
  grayscale_imags.append(rgb2gray(img)) 
  ax[i].imshow(img)
  ax[i].set_title('Orignal '+str(i))

fig.tight_layout()
plt.show()
```
